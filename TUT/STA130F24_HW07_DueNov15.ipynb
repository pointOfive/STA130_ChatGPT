#### A quick wrap up of Interaction

- **Definition:** An interaction occurs when the effect of one independent variable on the dependent variable depends on the level of another independent variable. It means that the impact of one variable changes based on the presence or magnitude of another variable.

- **Example**: Imagine you're making a smoothie. The taste of your smoothie depends on the ingredients you put in. Let's use two ingredients: bananas and strawberries.
    
#### Without Interaction
If there is no interaction, adding more bananas will always make the smoothie sweeter, no matter how many strawberries you put in. The effect of bananas on the smoothie is the same whether you add a lot of strawberries or just a few.

#### With Interaction
Interaction happens when the effect of adding bananas changes depending on how many strawberries you add.

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>
    
    
**R-squared (R²)**:<br>
**Meaning**: R² represents the proportion of the variance in the dependent variable that is explained by the independent variables. The value ranges from 0 to 1. A value closer to 1 indicates a stronger explanatory power of the model.<br>
**Purpose**: Used to evaluate the explanatory power of the model.<br><br>
    
**Adjusted R-squared (Adj. R²)**:<br>
**Meaning**: Adjusted R² adjusts for the number of predictors in the model, preventing the R² from increasing artificially by adding irrelevant variables. It only increases if the new predictor improves the model more than would be expected by chance.<br>
**Purpose**: Provides a more accurate measure of the model's explanatory power, especially when comparing models of different complexities.<br><br>
    
**F-statistic**:<br>
**Meaning**: The F-statistic tests the overall significance of the model, i.e., whether at least one of the predictors is significantly related to the dependent variable.<br>
**Purpose**: Used to assess the overall significance of the model. A higher F-statistic and a lower p-value indicate that the model is more significant.<br><br>
    
**Prob (F-statistic)**:<br>
**Meaning**: This is the p-value corresponding to the F-statistic. It indicates the probability that the observed F-statistic would occur if the null hypothesis were true.<br>
**Purpose**: Used to determine the overall significance of the model. A p-value less than a threshold (e.g., 0.05) indicates that the model is statistically significant.<br><br>

**AIC (Akaike Information Criterion)**:<br>
**Meaning**: AIC is used for model selection, balancing the goodness of fit and the complexity of the model. Lower AIC values indicate a better model.<br>
**Purpose**: Used to compare multiple models. The model with the lowest AIC is considered the best.<br><br>

**BIC (Bayesian Information Criterion)**:<br>
**Meaning**: BIC is similar to AIC but imposes a stricter penalty for model complexity. Lower BIC values indicate a better model.<br>
**Purpose**: Used to compare multiple models. The model with the lowest BIC is considered the best.<br><br>

**t-statistic (t)**:<br>
**Meaning**: The t-statistic tests whether a coefficient is significantly different from zero. It is calculated as the coefficient divided by its standard error.<br>
**Purpose**: Used to determine the significance of individual coefficients.

```python
from sklearn.model_selection import train_test_split
songs_training_data,songs_testing_data = train_test_split(songs, train_size=31)

reps = 100
in_sample_Rsquared = np.array([0.0]*reps)
out_of_sample_Rsquared = np.array([0.0]*reps)
for i in range(reps):
    songs_training_data,songs_testing_data = train_test_split(songs, train_size=31)
    final_model_fit = smf.ols(formula='danceability ~ energy * loudness + energy * mode', 
                              data=songs_training_data).fit()
    in_sample_Rsquared[i] = final_model_fit.rsquared
    out_of_sample_Rsquared[i] = np.corrcoef(songs_testing_data.danceability, 
                                            final_model_fit.predict(songs_testing_data))[0,1]**2
    
fig = px.scatter(pd.DataFrame({"In Sample Performance (Rsquared)": in_sample_Rsquared,
                              "Out of Sample Performance (Rsquared)": out_of_sample_Rsquared}), 
                 x="In Sample Performance (Rsquared)", y="Out of Sample Performance (Rsquared)")
fig.add_trace(go.Scatter(x=[0,1], y=[0,1], name="y=x", line_shape='linear'))    
```
